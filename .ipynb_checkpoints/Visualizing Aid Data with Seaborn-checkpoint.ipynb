```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style = "darkgrid")
# Sets style of the plots produced with a nice set of lines behind them

aid_df = pd.read_csv("aiddata-countries-only.csv", index_col=0)
%matplotlib inline
```

```python
#Looking at some characteristics of the data
print(aid_df.shape, end="\n \n")
print(aid_df.size, end="\n \n")
print(aid_df.info(), end="\n \n")
print(aid_df.head())
```

```python
# Pulling the 10 donors with the most instances in the Donor column
donors_top_10 = aid_df["donor"].value_counts().head(10)
donors_top_10
```

```python
# Plotting the top 10 donors
ax_donors = sns.barplot(x = donors_top_10.index, y = donors_top_10.values)
ax_donors.set(xlabel = "Country", ylabel = "Number of Donations")
plt.xticks(rotation = 45)
plt.title("Top 10 Aid Donors by Number of Donations")
plt.show()
```

```python
# Pulling the 10 recipients with the most instances in the Recipients column
recipients_top_10 = aid_df["recipient"].value_counts().head(10)
```

```python
# Plotting the top 10 recipients
ax_recipients = sns.barplot(x = recipients_top_10.index, y = recipients_top_10.values)
ax_recipients.set(xlabel = "Country", ylabel = "Number of Donations Received")
plt.xticks(rotation = 45)
plt.title("Top 10 Aid Recipients by Number of Donations Received")
plt.show()
```

```python
# Creating the index for the series of all donor countries
all_donor_countries = aid_df.donor.unique()

# Creating the list of amounts donated by country
country_don_amt = []
for country in all_donor_countries:
    don_amt = aid_df.loc[aid_df["donor"] == country]['commitment_amount_usd_constant'].sum()
    country_don_amt.append(don_amt)
```

```python
# Sanity check
aid_df.loc[aid_df["donor"] == "Kuwait"]['commitment_amount_usd_constant'].sum()
```

```python
# Actually creating the series which ties together the donor countries with the amounts donated
donations_srs = pd.Series((country_don_amt), index = all_donor_countries)

# Making the numbers simpler by putting them in billion USD
donations_bl_srs = donations_srs / 1000000000

#Sorting the data and pulling only the top 10 donors by amount donated
donations_bl_srs = donations_bl_srs.sort_values(ascending = False).head(10)
```

```python
#Plotting the top 10 donors
ax_don_amt = sns.barplot(x = donations_bl_srs.index, y = donations_bl_srs)
ax_don_amt.set(xlabel = "Country", ylabel = "Total Donations in USD (billions)")
plt.xticks(rotation = 45)
plt.title("Top 10 Aid Donors by Amount Donated")
plt.show()
```

```python
# Creating the index for the series of all recipient countries
all_recipient_countries = aid_df.recipient.unique()

# Creating the list of amounts received by country
country_rec_amt = []
for country in all_recipient_countries:
    rec_amt = aid_df.loc[aid_df["recipient"] == country]['commitment_amount_usd_constant'].sum()
    country_rec_amt.append(rec_amt)
```

```python
# Creating the series which ties the recipient countries to the amount received
receive_srs = pd.Series((country_rec_amt), index = all_recipient_countries)

# In billion USD
receive_bl_srs = receive_srs / 1000000000

# Sorting to pull only the top 10 recipient countries
receive_bl_srs = receive_bl_srs.sort_values(ascending = False).head(10)
```

```python
#Plotting the top 10 recipients
ax_rec_amt = sns.barplot(x = receive_bl_srs.index, y = receive_bl_srs)
ax_rec_amt.set(xlabel = "Country", ylabel = "Total Donations Received in USD (billions)")
plt.xticks(rotation = 45)
plt.title("Top 10 Aid Recipients by Amount Received")
plt.show()
```
